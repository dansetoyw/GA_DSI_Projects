## Background

The SAT and ACT are standardized tests that many colleges and universities in the United States require for their admissions process. This score is used along with other materials such as grade point average (GPA) and essay responses to determine whether or not a potential student will be accepted to the university.

The SAT has two sections of the test: Evidence-Based Reading and Writing and Math ([*source*](https://www.princetonreview.com/college/sat-sections)). The ACT has 4 sections: English, Mathematics, Reading, and Science, with an additional optional writing section ([*source*](https://www.act.org/content/act/en/products-and-services/the-act/scores/understanding-your-scores.html)). They have different score ranges, which you can read more about on their websites or additional outside sources (a quick Google search will help you understand the scores for each test):
* [SAT](https://collegereadiness.collegeboard.org/sat)
* [ACT](https://www.act.org/content/act/en.html)

Standardized tests have long been a controversial topic for students, administrators, and legislators. Since the 1940's, an increasing number of colleges have been using scores from sudents' performances on tests like the SAT and the ACT as a measure for college readiness and aptitude ([*source*](https://www.minotdailynews.com/news/local-news/2017/04/a-brief-history-of-the-sat-and-act/)). Supporters of these tests argue that these scores can be used as an objective measure to determine college admittance. Opponents of these tests claim that these tests are not accurate measures of students potential or ability and serve as an inequitable barrier to entry. Lately, more and more schools are opting to drop the SAT/ACT requirement for their Fall 2021 applications ([*read more about this here*](https://www.cnn.com/2020/04/14/us/coronavirus-colleges-sat-act-test-trnd/index.html)).

1. Manually calculate mean:

    Write a function that takes in values and returns the mean of the values. Create a list of numbers that you test on your function to check to make sure your function works!
    
    *Note*: Do not use any mean methods built-in to any Python libraries to do this! This should be done without importing any additional libraries.

```python
# Create function to calculate mean

# Below function is written without using any functions in python.

def mean(numbers):
    total = 0  # variable to store the sum of the numbers
    count = 0  # variable to store the counts of the numbers
    for number in numbers:
        total += number
        count += 1
    return total/count

# If we can use the sum or len functions in python, below will be the function.
# def mean(numbers):
#    return sum(numbers)/len(numbers)

test_num = [1,2,3,4,5]
mean(test_num)
```

2. Manually calculate standard deviation:

    The formula for standard deviation is below:

    $$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

    Where $x_i$ represents each value in the dataset, $\mu$ represents the mean of all values in the dataset and $n$ represents the number of values in the dataset.

    Write a function that takes in values and returns the standard deviation of the values using the formula above. Hint: use the function you wrote above to calculate the mean! Use the list of numbers you created above to test on your function.
    
    *Note*: Do not use any standard deviation methods built-in to any Python libraries to do this! This should be done without importing any additional libraries.

```python
# Create function to calculate standard deviation

def std(numbers):
    sum_sq_dev = 0 # variable to store the squared values of each deviation from the mean
    count = 0      # variable to store the counts of the numbers, assuming cant use len function. 
    for number in numbers:
        sum_sq_dev += (number-mean(numbers))**2
        count += 1
    return (sum_sq_dev/count)**0.5

std(test_num)
```

```python
# Create function to clean name
# lowercase the name
# replace space in name with underscore

def clean_name(name):
    return name.lower().replace(' ','_')
```

```python
# Create function to count the number of values that are not within range
# To check that all values are within range

def count_invalid_value(numbers, min_num, max_num): #parameters = (list of numbers, minimal value, maximal value)
    count = 0
    for number in numbers:
        if (number>max_num) or (number<min_num):
            count += 1
    if count == 0:
        print("No invalid value")
    else:
        print(f"There are {count} invalid values")
```

```python
# Line graph of ACT participation for each US state over the years i.e. From 2017 to 2019

# Create functions to extract data for the respective rows and columns
def extractdatax(dictionary):
    x = []
    for key in dictionary.keys():
        x.append(key)
    return (x)

def extractdatay(dictionary):
    y = []
    for value in dictionary.values():
        y.append(value)
    return (y)

# Obtain standard deviation of the ACT participation for year 2017 to 2019 by state
# by filtering the dataset to obtain the columns with ACT participation, calculating the standard deviation by row i.e.by state,
# and rounding up to 2 decimal places.
std_act_pt = act_sat_pt_score[['act17_participation','act18_participation','act19_participation']].std(axis=1).round(decimals=2)

plt.figure(figsize = (30, 100))

for n in range(0,51):
    # Create dictionary for the respective participation percentage by year, then by state (where n is the row of the state)
    act_pt_trend = {
    2017:act_sat_pt_score['act17_participation'][n],
    2018:act_sat_pt_score['act18_participation'][n],
    2019:act_sat_pt_score['act19_participation'][n],
    }


    plt.subplot(13,4,(n+1))  # chart arranged by 13 rows and 4 columns
    x = extractdatax(act_pt_trend)  # extract the years
    y = extractdatay(act_pt_trend)  # extract the participation percentage by state for respective years
    ax = sns.lineplot(x,y)
    ax.set_title(f"ACT Participation Percentage for {act_sat_pt_score['state'][n]} (std = {std_act_pt[n]})")
    ax.set_xlabel('years')
    ax.set_xticks(range(2017,2020))
    ax.set_ylabel(f"Participation (percentage in decimals)")

plt.show()
```

##### Observation for line graph of ACT participation for each US state over the years i.e. from 2017 to 2019

26 out of the 51 states had more than 5% decrease in the ACT participation from 2017 to 2019. 

Out of these 26 states, 3 states had more than 25% decrease in ACT participation.
- Colorado from 100% in 2017 to 27% in 2019 - down by 73%
- Illinois from 93% in 2017 to 35% in 2019 - down by 58%
- Alaska from 65% in 2017 to 38% in 2019 - down by 27%



3 states had more than 10% increase in ACT participation from 2017 to 2019.
- Ohio from 75% in 2017 to 100% in 2019 - up by 25%
- Nebraska from 84% in 2017 to 100% in 2019 - up by 16%
- Arizona from 62% in 2017 to 73% in 2019 - up by 11%

The ACT participation for the rest of the states remained relatively consistent over the years (with values within 3%). 

```python
# Line graph of ACT average composite in each US state over the years i.e. From 2017 to 2019

# Create functions to extract the respective rows and columns
def extractdatax(dictionary):
    x = []
    for key in dictionary.keys():
        x.append(key)
    return (x)

def extractdatay(dictionary):
    y = []
    for value in dictionary.values():
        y.append(value)
    return (y)

# Obtain standard deviation of the ACT average composite for year 2017 to 2019 by state
# by filtering the dataset to obtain the columns with composite, calculating the standard deviation and rounding up to 2 decimal places.
std_com = act_sat_pt_score.filter(regex='composite').std(axis=1).round(decimals=2)

plt.figure(figsize = (30, 70))

for n in range(0,51):
    act_com_trend = {
    2017:act_sat_pt_score['act17_composite'][n],
    2018:act_sat_pt_score['act18_composite'][n],
    2019:act_sat_pt_score['act19_composite'][n],
    }


    plt.subplot(11,5,(n+1))
    x = extractdatax(act_com_trend)
    y = extractdatay(act_com_trend)
    ax = sns.lineplot(x,y)
    ax.set_title(f"Average Composite for {act_sat_pt_score['state'][n]} (std = {std_com[n]})")
    ax.set_xlabel('years')
    ax.set_xticks(range(2017,2020))
    ax.set_ylabel('Average Composite')

plt.show()
```
