## Background

The SAT and ACT are standardized tests that many colleges and universities in the United States require for their admissions process. This score is used along with other materials such as grade point average (GPA) and essay responses to determine whether or not a potential student will be accepted to the university.

The SAT has two sections of the test: Evidence-Based Reading and Writing and Math ([*source*](https://www.princetonreview.com/college/sat-sections)). The ACT has 4 sections: English, Mathematics, Reading, and Science, with an additional optional writing section ([*source*](https://www.act.org/content/act/en/products-and-services/the-act/scores/understanding-your-scores.html)). They have different score ranges and the details can be found in below link.
* [SAT](https://collegereadiness.collegeboard.org/sat)
* [ACT](https://www.act.org/content/act/en.html)

Standardized tests have long been a controversial topic for students, administrators, and legislators. Since the 1940's, an increasing number of colleges have been using scores from sudents' performances on tests like the SAT and the ACT as a measure for college readiness and aptitude ([*source*](https://www.minotdailynews.com/news/local-news/2017/04/a-brief-history-of-the-sat-and-act/)). Lately, more and more schools are opting to drop the SAT/ACT requirement for their applications ([*source*](https://www.cnn.com/2020/04/14/us/coronavirus-colleges-sat-act-test-trnd/index.html)).

```python
# Create function to calculate mean

# Below function is written without using any functions in python.

def mean(numbers):
    total = 0  # variable to store the sum of the numbers
    count = 0  # variable to store the counts of the numbers
    for number in numbers:
        total += number
        count += 1
    return total/count

# If we can use the sum or len functions in python, below will be the function.
# def mean(numbers):
#    return sum(numbers)/len(numbers)

test_num = [1,2,3,4,5]
mean(test_num)
```

2. Manually calculate standard deviation:

    The formula for standard deviation is below:

    $$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

    Where $x_i$ represents each value in the dataset, $\mu$ represents the mean of all values in the dataset and $n$ represents the number of values in the dataset.

    Write a function that takes in values and returns the standard deviation of the values using the formula above. Hint: use the function you wrote above to calculate the mean! Use the list of numbers you created above to test on your function.
    
    *Note*: Do not use any standard deviation methods built-in to any Python libraries to do this! This should be done without importing any additional libraries.

```python
# Create function to calculate standard deviation

def std(numbers):
    sum_sq_dev = 0 # variable to store the squared values of each deviation from the mean
    count = 0      # variable to store the counts of the numbers, assuming cant use len function. 
    for number in numbers:
        sum_sq_dev += (number-mean(numbers))**2
        count += 1
    return (sum_sq_dev/count)**0.5

std(test_num)
```

```python
# Create function to count the number of values that are not within range
# To check that all values are within range

def count_invalid_value(numbers, min_num, max_num): #parameters = (list of numbers, minimal value, maximal value)
    count = 0
    for number in numbers:
        if (number>max_num) or (number<min_num):
            count += 1
    if count == 0:
        print("No invalid value")
    else:
        print(f"There are {count} invalid values")
```

```python
# Line graph of average ACT and SAT participation over the years i.e. From 2017 to 2019

# Create functions to extract data for the respective rows and columns
def extractdatax(dictionary):
    x = []
    for key in dictionary.keys():
        x.append(key)
    return (x)

def extractdatay(dictionary):
    y = []
    for value in dictionary.values():
        y.append(value)
    return (y)

plt.figure(figsize = (20, 5))

for n in range(0,2):
    # Create dictionary for the respective participation percentage of by test type and by year
    act_sat_pt_trend = {
    0 : {2017:act_sat_pt_score['act17_participation'].mean(),
        2018:act_sat_pt_score['act18_participation'].mean(),
        2019:act_sat_pt_score['act19_participation'].mean()
        },
    1 : {2017:act_sat_pt_score['sat17_participation'].mean(),
        2018:act_sat_pt_score['sat18_participation'].mean(),
        2019:act_sat_pt_score['sat19_participation'].mean()
        }   
    }


    plt.subplot(1,3,(n+1))  # chart arranged by 1 row and 3 columns
    x = extractdatax(act_sat_pt_trend[n])  # extract the years
    y = extractdatay(act_sat_pt_trend[n])  # extract the average participation percentage for respective years
    ax = sns.lineplot(x,y)
    ax.set_title(f"Average Participation Percentage for {['ACT','SAT'][n]}")
    ax.set_xlabel('Years')
    ax.set_xticks(range(2017,2020))
    ax.set_ylim(0,1)
    ax.set_ylabel(f"Participation (percentage in decimals)")
    for xl,yl in zip(x,y):   # To put data label

        label = "{:.2f}".format(yl)

        plt.annotate(label, # this is the text
                     (xl,yl), # these are the coordinates to position the label
                     textcoords="offset points", # how to position the text
                     xytext=(0,10), # distance from text to points (x,y)
                     ha='center') # horizontal alignment can be left, right or center
    
plt.show()

# Please see charts after the warning messages
```

##### Observation for line graph of average ACT and SAT participation over the years i.e. From 2017 to 2019

ACT participation percentage decreased from 65% in 2017 to 59% in 2019.

SAT participation percentage increased from 40% in 2017 to 49% in 2019.
