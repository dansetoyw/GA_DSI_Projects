## Background

Real Sky Estate Development has been developing residential areas around Ames and is looking to procure and develop new housing in the area. 

Previous housing developments were not returning favourable profits due to the recent pandemic and political tensions. It also caused a rise in the costs of living and building materials over the years.

Facing a forecasted recession in the upcoming year, Real Sky senior management team has reached out to our data science team to pinpoint factors that will direct towards revamping the company’s focus structure to improve the attractiveness and sales price of new housing developments.  

We will be using data obtained from Ames, Iowa Assessor's Office on residential properties sold in Ames, IA from 2006 to 2010.

### Data Source

Source: Ames, Iowa Assessor’s Office 

* [`train.csv`](../data/train.csv): Data set contains information from the Ames Assessor’s Office used in computing assessed values for individual residential properties sold in Ames, IA from 2006 to 2010.


### Outside Research

Based on ['Forbes'](https://www.forbes.com/sites/forbesrealestatecouncil/2020/04/27/15-most-desirable-home-features-to-todays-buyers/?sh=425af7c97a4e) website, the 15 most desirable home features are community amenities, location, great schools in the area, flexible space, updated kitchen, a large yard, high-end finishes, garage, smart home technology, open living space, logical design and layout, legal flexibility, convenience, accessory dwelling units and value. Some of these features are provided in the data file. 

Based on ['Neighborhoodscout'](https://www.neighborhoodscout.com/ia/ames) website, north-west of Ames has lower crime rate and better education rating.



## Data Import and Cleaning

### Import Libraries and set display settings

```python
## Import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import seaborn as sns

from statistics import mode, median
```

```python
## Change display settings

# Change the display of dataframe - max number of columns and max number of rows
pd.set_option('display.max_columns',90)
pd.set_option('display.max_rows',60)
```

```python
## Import dataset and format column names to lowercase as well as space replace by underscore

# Import training dataset on Ames Housing and clean column names

prop_trn = pd.read_csv('../data/train.csv')
prop_trn.columns = prop_trn.columns.str.lower().str.replace(' ','_')
```

```python
# Summary of the training dataset on Ames Housing

prop_trn.info()
prop_trn.head()
```

```python
### Cleaning lot frontage ###

## Fill na with median value
prop_trn.lot_frontage.fillna(median(prop_trn.lot_frontage), inplace = True)
```

```python
### Cleaning alley predictor ###

## Based on the data dictionary provided, NA for 'alley' predictor refers to 'No alley access'
## Change NA to 'none' to avoid error
prop_trn.loc[prop_trn[prop_trn.alley.isnull()].index, ['alley']] = 'none'

# To consider to drop this column as majority has no alley access
```

```python
### Cleaning masonry veneer predictors ###

# Fill in NA with the mode for mas_vnr_type (which is 'None')
# as there was no obvious relationship with other features (e.g. exterior_1st, exterior_2nd etc)
prop_trn.mas_vnr_type.fillna(mode(prop_trn.mas_vnr_type), inplace = True)

# Since the value for mas_vnr_type is 'None', the area is zero.
prop_trn.mas_vnr_area.fillna(0.0, inplace = True)

# To consider to drop the rows with na instead of imputation as there were only 22 rows with NA.
```

```python
### Cleaning basement predictors ###

## Based on the data dictionary provided, NA for 'bsmt_' predictors refers to 'No Basement'
## Change NA to 'none' to avoid error
prop_trn.loc[prop_trn[prop_trn.bsmt_qual.isnull()].index, ['bsmt_qual']] = 'none'
prop_trn.loc[prop_trn[prop_trn.bsmt_cond.isnull()].index, ['bsmt_cond']] = 'none'
prop_trn.loc[prop_trn[prop_trn.bsmtfin_type_1.isnull()].index, ['bsmtfin_type_1']] = 'none'

## Indexes 1456, 1547, 1997 have missing values for basement exposure becuase they have basement values for other columns. 
# Change NA for these indexes to 'none' assuming no exposure because unable to imputate based on other predictors.
prop_trn.loc[prop_trn[(prop_trn.bsmt_exposure.isnull()) & (prop_trn.bsmt_qual != 'none')].index, ['bsmt_exposure']] = 'none'
# Change NA to 'none' to avoid error for 'bsmt_exposure' since there is no basement for the rest.
prop_trn.loc[prop_trn[prop_trn.bsmt_exposure.isnull()].index, 'bsmt_exposure'] = 'none'

## Index 1147 has missing value for bsmtfin_type_2 becuase it has value for 'bsmtfin_sf_2'
# replace NA with the mode of the values in 'bsmtfin_type_2' because unable to imputate based on other predictors
prop_trn.loc[prop_trn[(prop_trn.bsmtfin_type_2.isnull()) & (prop_trn.bsmt_qual != 'none')].index,
             'bsmtfin_type_2'] = mode(prop_trn.bsmtfin_type_2)
# Change NA to 'none' to avoid error for 'bsmtfin_type_2' since there is no basement for the rest
prop_trn.loc[prop_trn[prop_trn.bsmtfin_type_2.isnull()].index, 'bsmtfin_type_2'] = 'none'

## Change NA to 0.0 for index 1327 with assumption that it has no basement as it is consistent with the rest of the columns of 
## same row
prop_trn.loc[prop_trn[prop_trn.bsmtfin_sf_1.isnull()].index, 'bsmtfin_sf_1'] = 0.0
prop_trn.loc[prop_trn[prop_trn.bsmtfin_sf_2.isnull()].index, 'bsmtfin_sf_2'] = 0.0
prop_trn.loc[prop_trn[prop_trn.bsmt_unf_sf.isnull()].index, 'bsmt_unf_sf'] = 0.0
prop_trn.loc[prop_trn[prop_trn.total_bsmt_sf.isnull()].index, 'total_bsmt_sf'] = 0.0

# Have verified that ((prop_trn.bsmtfin_sf_1 + prop_trn.bsmtfin_sf_2 + prop_trn.bsmt_unf_sf) == prop_trn.total_bsmt_sf) is True 
# for all columns.

## Change NA to 0.0 for indexes 616 & 1327 for bsmt full/half bath since there is no basement. 
prop_trn.loc[prop_trn[prop_trn.bsmt_full_bath.isnull()].index, 'bsmt_full_bath'] = 0.0
prop_trn.loc[prop_trn[prop_trn.bsmt_half_bath.isnull()].index, 'bsmt_half_bath'] = 0.0
```

```python
### Cleaning fireplaces ###

## Based on the data dictionary provided, NA for 'fireplace_qu' refers to 'No fireplace'
## Change NA to 'none' to avoid error
prop_trn.loc[prop_trn[prop_trn.fireplace_qu.isnull()].index, 'fireplace_qu'] = 'none'
```

```python
### Cleaning garage predictors ###

## Based on the data dictionary provided, NA for 'garage_type' refers to 'No garage'
## Change NA to 'none' to avoid error
prop_trn.loc[prop_trn[prop_trn.garage_type.isnull()].index, 'garage_type'] = 'none'

## Index 1712 has missing values for all garage predictors except for the 'garage_type'. 
# Garage type was indicated as 'Detchd'. 
# Fill in the missing values with detached type garage's median when the value is number and mode when the value is string.
prop_trn.loc[[1712],['garage_finish']] = mode(prop_trn[prop_trn.garage_type == 'Detchd']['garage_finish'])
prop_trn.loc[[1712],['garage_cars']] = median(prop_trn[prop_trn.garage_type == 'Detchd']['garage_cars'])
prop_trn.loc[[1712],['garage_area']] = median(prop_trn[prop_trn.garage_type == 'Detchd']['garage_area'])
prop_trn.loc[[1712],['garage_qual']] = mode(prop_trn[prop_trn.garage_type == 'Detchd']['garage_qual'])
prop_trn.loc[[1712],['garage_cond']] = mode(prop_trn[prop_trn.garage_type == 'Detchd']['garage_cond'])
# Fill in the garage built year with house built year 
# with assumption that garage was built together with house in the same year.
prop_trn.loc[[1712],['garage_yr_blt']] = prop_trn.loc[[1712],['year_built']].values 

## Change NA to 'none' to avoid error for the rest of the column with NA since there is no garage for these rows
prop_trn.loc[prop_trn[prop_trn.garage_finish.isnull()].index, 'garage_finish'] = 'none'
prop_trn.loc[prop_trn[prop_trn.garage_qual.isnull()].index, 'garage_qual'] = 'none'
prop_trn.loc[prop_trn[prop_trn.garage_cond.isnull()].index, 'garage_cond'] = 'none'

# Fill in NA with 0 for garage year built for those without garage built to avoid error
prop_trn.garage_yr_blt.fillna(0, inplace = True)
# To consider dropping this column since it is likely to be a weak predictor and not relevant
```

```python
### Cleaning pool predictor ###

## Based on the data dictionary provided, NA for 'garage_qc' refers to 'No pool'
## Change NA to 'none' to avoid error
prop_trn.loc[prop_trn[prop_trn.pool_qc.isnull()].index, 'pool_qc'] = 'none'
```

```python
### Cleaning fence predictor ###

## Based on the data dictionary provided, NA for 'fence' refers to 'No fence'
## Change NA to 'none' to avoid error
prop_trn.loc[prop_trn[prop_trn.fence.isnull()].index, 'fence'] = 'none'
```

```python
### Cleaning misc feature predictor ###

## Based on the data dictionary provided, NA for 'misc_feature' refers to 'None'
## Change NA to 'none' to avoid error
prop_trn.loc[prop_trn[prop_trn.misc_feature.isnull()].index, 'misc_feature'] = 'none'
```

```python
### Cleaning 'yr_sold' ###

## Indexes 1719 and 1885 have 'yr_sold' before 'year_built' or 'year_remod/add'. 
# Likely to be typo because the year that the house was sold should be after it is built/remodelled as 'sale_type' = 'New'
# Change the year_sold to be the same year as the year_built/remod assuming it is sold after remod

prop_trn.loc[[1719], ['yr_sold']] = prop_trn.loc[[1719], ['year_remod/add']].values
prop_trn.loc[[1885], ['yr_sold']] = prop_trn.loc[[1885], ['year_remod/add']].values

# Checked that all values of 'yr_sold' are between 2006 and 2010 (inclusive).
```

```python
### Change the columns data type ###

prop_trn = prop_trn.astype({
    'bsmt_full_bath' : 'int64',
    'bsmt_half_bath' : 'int64',
    'garage_cars'    : 'int64',
    'bsmtfin_sf_1'   : 'int64',
    'bsmtfin_sf_2'   : 'int64',
    'bsmt_unf_sf'    : 'int64',
    'total_bsmt_sf'  : 'int64',
    'garage_area'    : 'int64'
})
```

```python
# Create an empty dataframe to store the ordinal features with quality values converted to score
feat_ord_score_df = pd.DataFrame()
```

```python
# Create a dictionary for conversion of quality description to score (with excellent having the highest score)
# with assumption that the differences between the quality description are equal
# Ex	Excellent       = 5
# Gd	Good            = 4
# TA	Typical/Average = 3
# Fa	Fair            = 2
# Po	Poor            = 1
# None	None            = 0

qual_score_dict = {
    'none' : 0,
    'Po'   : 1,
    'Fa'   : 2,
    'TA'   : 3,
    'Gd'   : 4,
    'Ex'   : 5
}
```

```python
# List of ordinal features with standard quality description i.e. can use qual_score_dict for conversion
feat_ord_standard_qual = ['exter_qual', 'exter_cond', 'bsmt_qual',
                          'bsmt_cond',  'heating_qc', 'kitchen_qual', 
                          'fireplace_qu',  'garage_qual',  'garage_cond', 'pool_qc']

# Convert quality description to score for ordinal features with standard description 
# and add them to the 'feat_ord_score_df' dataframe 
for feature in feat_ord_standard_qual:
    feat_ord_score_df[feature + '_score'] = prop_trn[feature].map(qual_score_dict)
```

```python
# Convert quality description of the rest of the ordinal features to score
# with assumption that the differences between the quality description are equal
# and add them to the 'feat_ord_score_df' dataframe 

feat_ord_score_df['lot_shape_score'] = prop_trn['lot_shape'].map({'Reg':4, 'IR1':3, 'IR2':2, 'IR3':1})

feat_ord_score_df['utilities_score'] = prop_trn['utilities'].map({'AllPub':4, 'NoSewr':3, 'NoSeWa':2, 'ELO':1})

feat_ord_score_df['land_slope_score'] = prop_trn['land_slope'].map({'Gtl':3, 'Mod':2, 'Sev':1})

feat_ord_score_df['bsmt_exposure_score'] = prop_trn['bsmt_exposure'].map({'Gd':4, 'Av':3, 'Mn':2, 'No':1, 'none':0})

feat_ord_score_df['bsmtfin_type_1_score'] = prop_trn['bsmtfin_type_1'].map({'GLQ':6, 'ALQ':5, 'BLQ':4, 'Rec':3, 
                                                                            'LwQ':2, 'Unf':1, 'none':0})

feat_ord_score_df['bsmtfin_type_2_score'] = prop_trn['bsmtfin_type_2'].map({'GLQ':6, 'ALQ':5, 'BLQ':4, 'Rec':3, 
                                                                            'LwQ':2, 'Unf':1, 'none':0})

feat_ord_score_df['electrical_score'] = prop_trn['electrical'].map({'SBrkr':5, 'FuseA':4, 'FuseF':3, 'FuseP':2, 'Mix':1})

feat_ord_score_df['functional_score'] = prop_trn['functional'].map({'Typ':8, 'Min1':7, 'Min2':6, 'Mod':5, 'Maj1':4, 
                                                                    'Maj2':3, 'Sev':2, 'Sal':1})

feat_ord_score_df['garage_finish_score'] = prop_trn['garage_finish'].map({'Fin':3, 'RFn':2, 'Unf':1, 'none':0})

feat_ord_score_df['paved_drive_score'] = prop_trn['paved_drive'].map({'Y':3, 'P':2, 'N':1})

feat_ord_score_df['fence_score'] = prop_trn['fence'].map({'GdPrv':4, 'MnPrv':3, 'GdWo':2, 'MnWw':1, 'none':0})
```

```python
## Heatmap of the ordinal features based on scores and sale price

ord_feat_w_saleprice = pd.concat([feat_ord_score_df, prop_trn['saleprice']], axis = 1)

# np.zeros_like returns an array of zeros with the same shape and type as X-data
mask = np.zeros_like(ord_feat_w_saleprice.corr())

# np.triu_indices_from(mask) --> returns the indices for the upper-triangle of corr matrix.
mask[np.triu_indices_from(mask)] = True

plt.figure(figsize = (20,20))
sns.heatmap(ord_feat_w_saleprice.corr(),
            annot = True,
           cmap = 'coolwarm',
           mask = mask,
           vmin = -1,
           vmax = 1)
plt.title('Heatmap of ordinal features and saleprice', fontsize = 30)

plt.show()
```

##### Interpretation
- Features with more than 0.6 correlation coefficient with saleprice are exter_qual, bsmt_qual and kitchen_qual.
- These features are potential features for modelling.

```python
# Scatterplot of ordinal features against saleprice

plt.figure(figsize = (20,40))

ord_feat_names = feat_ord_score_df.columns

for n in range(0,21):
   
    plt.subplot(7,3,n+1)
    
    r_score = round(stats.pearsonr(ord_feat_w_saleprice[ord_feat_names[n]], prop_trn.saleprice)[0],2)
    
    plt.scatter(ord_feat_w_saleprice[ord_feat_names[n]], prop_trn.saleprice, s=8)
    plt.xlabel(f'{ord_feat_names[n]}');
    plt.ylabel('Sale Price (USD)');
    plt.title(f"Scatter Plot of Sale Price vs {ord_feat_names[n]} (r={r_score})")    
    
    #find the best fit line
    a, b = np.polyfit(ord_feat_w_saleprice[ord_feat_names[n]], prop_trn.saleprice, 1)
    #Add best fit line to plot
    plt.plot(ord_feat_w_saleprice[ord_feat_names[n]],a*(ord_feat_w_saleprice[ord_feat_names[n]])+b, color='red')

plt.show()
```

##### Interpretation
- Features with more than 0.6 correlation coefficient with saleprice have positive correlation with saleprice. 
- The rest of the features have relatively poorer correlation with saleprice.

```python
## Get Correlation coefficient for each variable in each of the nominal features column with saleprice

# Dummify the nominal features
dummified_nom_feat_df = pd.get_dummies(columns=prop_trn_feat_nom , data=prop_trn[prop_trn_feat_nom])

# Concatenate the dummified nominal features dataframe and saleprice
# Get correlation coefficient of the sub-features with saleprice
# Sort it according to descending order
# Get 10 top correlation coefficient of sub-features with saleprice
pd.concat([dummified_nom_feat_df, prop_trn['saleprice']], axis = 1).corr()['saleprice'].sort_values(ascending = False).head(11)
```

##### Interpretation
- None of the variable in each of the nominal features column has higher than 0.6 correlation coefficient. Hence, unlikely to add them to the modelling.

Further examination of selected features.

```python
# Selected features for modelling

sel_features = ['overall_qual','total_bsmt_sf','1st_flr_sf','gr_liv_area','garage_area','exter_qual','bsmt_qual',
                'kitchen_qual','bsmt_qual']

# Neighborhood is selected however, it is nominal features, hence, not included in the list.

# Get dataframe with sel_features

prop_trn_sel_features = prop_trn[sel_features]
```

- The selected features are reasonably normally distributed.
- The selected features' error term is reasonably same across all values. i.e. Homoscedasticity.
